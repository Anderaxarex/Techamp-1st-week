# Dictionary

Dictionaries are very useful, especially when you want to save items and objects with a specific key. For instance, it can be used to store the telephone numbers of your friends: "Marco" -> 33344556677. "Marco" in this case, is the key. Instead, its telephone number is the value.

If you want to create a dictionary initially empty, you can declare it with:

```python
my_dict = {}
```

To insert an entry in your dictionary, such as "Marco" -> 33344556677, you can do it in this way:

```python
my_dict["Marco"] = 33344556677
```

Now, the dictionary contains your entry. You can check it by printing the dictionary.

```python
print(my_dict)
```

We can instert as many entry we want and the dictionary will contain all of them.

```python
my_dict["Antonio"] = 33377889900
my_dict["Lorenzo"] = 33311882200
my_dict["Mario"] = 33343444546
```

```python
print(my_dict)
```

If we forget the telephone number of somebody, we can use the dictionary to remember it :D

```python
print(my_dict["Antonio"])
```

If we want to see all the names(keys) in the dictionary, we can see them with:

```python
print(my_dict.keys())
```

And we can also do a list of them! Maybe you want to iterate over all of them.

```python
my_keys_list = list(my_dict.keys())
print(my_keys_list)
```

And we can do the same with the telephone numbers:

```python
print(my_dict.values())
```

Moreover, we can also replace the telephone number of a person.

```python
my_dict["Marco"] = "Telephone Lost"
print(my_dict["Marco"])
```

### NB:
#### 1) A key can only have one value
#### 2) Both the key and the value can be of any type
